```python
#TASK 2

'''expected output [5,10,15,20] 
means want to get an expected list 
that contain kelipatan 5 dimulai dari angka setelah 1 dan berakir jika sudah mencapai20 '''

def find_multiple(start,end,multiple):
    listkosong = []
    for i in range(start, end+1):
        if i % multiple == 0:       # bagian ini masih mayan bingung apakah dia abis looping cek bakal masuk ke listkosong atau gimana
            listkosong.append(i)

    return listkosong

print(find_multiple(start=1,end=20,multiple=5))

print(find_multiple(start=1,end=50,multiple=10))
```

```python
#TASK 3
def grade_converter(convert_to,**kwargs):
    dictkosong = {}
    for username,score in kwargs.items():
        if convert_to.strip().lower()=="gpa":
                if score >= 85:
                    dictkosong[username]="4"
                if score >=70:
                    dictkosong[username]="3"
                elif score >=60:
                    dictkosong[username]="2.5"
                elif score >=50:
                    dictkosong[username]="2"
                elif score >=40:
                    dictkosong[username]="1.5"
                else:
                    dictkosong[username]="1"
        elif convert_to.strip().lower()=="grade":
                if score >= 85:
                    dictkosong[username]="A"
                if score >=70:
                    dictkosong[username]="B"
                elif score >=60:
                    dictkosong[username]="C"
                elif score >=50:
                    dictkosong[username]="D"
                elif score >=40:
                    dictkosong[username]="F"
                else:
                    dictkosong[username]="F"
                
    return dictkosong

print(grade_converter(convert_to="grade",albert=90,dwi=82,syahdan=38,veronica=50))

# {'albert':"A",'dwi':"B", 'syahdan':"F",'veronica':"D"}s
```

```python
#TASK 4 (belum selesai)

def fibonaciseq(n):
    seqlist = []
    a, b = 0 , 1

    for _ in range (n):
        seqlist.append(a)
        a,b= b, a+b
    return seqlist

num = 5
print(last:= fibonaciseq(num))
# fibonaciseq(5)

def fibonacci_number(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    
    return b

# Example usage:
n = 10
print(f"The {n}-th Fibonacci number is: {fibonacci_number(n)}")
```
