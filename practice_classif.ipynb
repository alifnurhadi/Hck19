```python
from sklearn.model_selection import train_test_split

# split feature and target
x = df.drop('target',axis=1)
y = df['target']
```

```python
# splitting data
x_train,x_test,y_train,y_test = train_test_split(x, y , test_size= 0.2 , random_state= 42)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
import numpy as np

def evaluate_k(classifier, k: int):
    pipeline = Pipeline([
        ('select', SelectKBest(f_classif, k=k)),
        ('classifier', classifier)
    ])
    scores = cross_val_score(pipeline, x_train, y_train, cv=5)
    return scores.mean()

# Define classifiers
classifiers = {
    "GaussianNB": GaussianNB(),
    "LogisticRegression": LogisticRegression()
}

# Try different k values
k_values = range(1, x_train.shape[1] + 1)

# Evaluate each classifier
for clf_name, clf in classifiers.items():
    scores = [evaluate_k(clf, k) for k in k_values]
    best_k = k_values[np.argmax(scores)]
    print(f"Best k for {clf_name}: {best_k}")

    # Refit with best k
    select_kbest = SelectKBest(f_classif, k=best_k)
    select_kbest.fit(x_train, y_train)

    selected_columns = x_train.columns[select_kbest.get_support()]
    print(f"Selected columns for {clf_name}: {selected_columns}")
```

```python
# feature selection

selection:list = [x_train,x_test]

for select in selection :
    select.drop([ 'chol', 'fbs'],axis=1 ,inplace=True)
```

```python
# scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)

x_train_scale = scaler.transform(x_train)
x_test_scale = scaler.transform(x_test)
```

```python
# modeling with logistic regression.

from sklearn.metrics import accuracy_score, classification_report

log_reg = LogisticRegression()

log_reg.fit(x_train_scale,y_train)

y_pred = log_reg.predict(x_test_scale)

accuracy = accuracy_score(y_pred,y_test)

print(f"Accuracy: {accuracy:.4f}")

# Detailed classification report
print(classification_report(y_test, y_pred))
```
